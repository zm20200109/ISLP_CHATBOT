```python
import sys
sys.path.append("/content/drive/MyDrive/project_work/code/")
from chroma_db_connection import create_retriever

chunks_path = "/content/drive/MyDrive/project_work/code/results/chunked_elements_1.pkl"#"results/chunked_elements_1.pkl" # chunk_by_title strategy, combine_text_under_n_chars=1000,max_characters=1600, multipage_sections=True, new_after_n_chars=1200, overlap=True,
txt_summaries_path = "/content/drive/MyDrive/project_work/code/results/1_txt_summaries.pkl"#"results/1_txt_summaries.pkl"
tbl_summaries_path = "/content/drive/MyDrive/project_work/code/results/1_table_summaries.pkl"#"results/1_table_summaries.pkl"
retriever = create_retriever(chunks_path,txt_summaries_path,tbl_summaries_path,"similarity",5)
```

```python
from multiquery_chain import final_chain
from evaluation import Evaluation
```

```python
def caluclate_avg_metrics_on_dataset(dataset, chain_name):
  mean_context_prec = dataset["context_precision"].mean()
  mean_context_rec = dataset["context_recall"].mean()
  mean_faithfulness = dataset["faithfulness"].mean()
  mean_relevancy = dataset["answer_relevancy"].mean()
  return {"chain_name":chain_name, "mean_context_precision":mean_context_prec,"mean_context_recall":mean_context_rec,"mean_faithfulness":mean_faithfulness, "mean_relevancy":mean_relevancy}
```

**Evaluacija MultiQuery query translation mehanizma**

```python
multiquery_chain = final_chain(retriever)
```

```python
eval_multiquery = Evaluation(retriever,multiquery_chain)
multiqery_results_df = eval_multiquery.evaluate_chain()
```

```python
 # context_precision, context_recall, faithfulness, answer_relevancy
avg_metrics_multiquery = caluclate_avg_metrics_on_dataset(multiqery_results_df,"multiquery_chain")
```

```python
avg_metrics_multiquery
```

```python
multiqery_results_df.to_csv("/content/results/multiquery_eval_metrics.csv")
```
